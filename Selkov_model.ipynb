```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 25})
```

```python
def ode(z, t, v, w):
    x, y = z
    dzdt = [v - x * y ** 2, x * y ** 2 - w * y]
    return dzdt
```

```python
def calcODE(args, z0, dz0, ts = 10, nt = 100):
    z0 = [z0, dz0]
    t = np.linspace(0, ts, nt)
    sol = odeint(ode, z0, t, args)
    return sol
```

```python
def draw(args, x0, y0, ts=10, nt=100):
    sol = calcODE(args, x0, y0, ts, nt)
    plt.figure(figsize=(10, 10))
    plt.plot(sol[:, 0], sol[:, 1], 'g')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(10, 10))
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(np.linspace(0, ts, nt), sol[:, 0], 'b')
    plt.plot(np.linspace(0, ts, nt), sol[:, 1], 'r')

    plt.xlabel('t')
    plt.ylabel('x, y')
    plt.grid()
    plt.show()
```

```python
def multidraw(args, x0, y0, ts=10, nt=101):

    sol = calcODE(args, x0+0.1, y0, ts, nt)
    plt.figure(figsize=(10, 10))
    plt.plot(sol[:, 0], sol[:, 1], 'r')
    
    sol = calcODE(args, x0 - 0.1, y0, ts, nt)
    plt.plot(sol[:, 0], sol[:, 1], 'g')
    
    sol = calcODE(args, x0, y0+0.1, ts, nt)
    plt.plot(sol[:, 0], sol[:, 1], 'b')
    
    sol = calcODE(args, x0, y0-0.1, ts, nt)
    plt.plot(sol[:, 0], sol[:, 1], 'c')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()

    plt.figure(figsize=(10, 10))
    sol = calcODE(args, x0+0.1e-2, y0+0.1e-2, ts, nt)
    plt.plot(np.linspace(0, ts, nt), sol[:, 0], 'b')
    plt.plot(np.linspace(0, ts, nt), sol[:, 1], 'r')

    plt.xlabel('t')
    plt.ylabel('x, y')
    plt.grid()
    plt.show()
```
