```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 25})
```

```python
def ode(z, t, v, w):
    x, y = z
    dzdt = [v - x * y ** 2, x * y ** 2 - w * y]
    return dzdt
```

```python
def calcODE(args, z0, dz0, ts = 10, nt = 100):
    z0 = [z0, dz0]
    t = np.linspace(0, ts, nt)
    sol = odeint(ode, z0, t, args)
    return sol
```

```python
def draw(args, x0, y0, ts=10, nt=100):
    sol = calcODE(args, x0, y0, ts, nt)
    plt.figure(figsize=(10, 10))
    plt.plot(sol[:, 1], sol[:, 0], 'g')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(10, 10))
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(np.linspace(0, ts, nt), sol[:, 0], 'b')
    plt.plot(np.linspace(0, ts, nt), sol[:, 1], 'r')

    plt.xlabel('t')
    plt.ylabel('x, y')
    plt.grid()
    plt.show()
```

```python
def multidraw(v, w, x0, y0, ts=10, nt=101):
    args = (v, w)
    sol = calcODE(args, x0, y0, ts, nt)
    plt.figure(figsize=(10, 10))
    plt.plot(sol[:, 1], sol[:, 0], 'r')
    
    args = (v + 0.1, w)
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(sol[:, 1], sol[:, 0], 'g')
    
    args = (v + 0.2, w)
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(sol[:, 1], sol[:, 0], 'b')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()

    plt.figure(figsize=(10, 10))
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(np.linspace(0, ts, nt), sol[:, 0], 'b')
    plt.plot(np.linspace(0, ts, nt), sol[:, 1], 'r')

    plt.xlabel('t')
    plt.ylabel('x, y')
    plt.grid()
    plt.show()
```

```python
v = 1 + 1e-2 # область II на диаграмме (учтойчивый фокус)
w = 1 ** (2/3)
args = (v, w)
draw(args, x0=2, y0=2, ts = 310, nt = 100000)
draw(args, x0=w*w / v + 1e-3, y0=v / w + 1e-3, ts = 300, nt = 100000) # прямо около стационара взяли точки

# multidraw(v, w, x0=3, y0=3, ts = 50, nt = 10000)

v = 1 - 1e-2 # область III на диаграмме (неустойчивый фокус)
w = 1 ** (2 / 3) 
args = (v, w)
draw(args, x0=2, y0=2, ts = 300, nt = 100000)
draw(args, x0=w*w / v + 1e-2, y0=v / w + 1e-2, ts = 500, nt = 100000) # прямо около стационара начальные точки
# multidraw(v, w, x0=3, y0=3, ts = 50, nt = 10000)
```
