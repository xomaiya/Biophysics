```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def ode(z, t, v, w):
    x, y = z
    dzdt = [v - x * y * y, x * y * y - w * y]
    return dzdt
```

```python
def calcODE(args, z0, dz0, ts = 10, nt = 100):
    z0 = [z0, dz0]
    t = np.linspace(0, ts, nt)
    sol = odeint(ode, z0, t, args)
    return sol
```

```python
def draw(args, x0, y0, ts=10, nt=100):
    sol = calcODE(args, x0, y0, ts, nt)
    plt.figure(figsize=(10, 10))
    plt.plot(sol[:, 1], sol[:, 0], 'g')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
```

```python
def multidraw(v, w, x0, y0, ts=10, nt=101):
    args = (v, w)
    sol = calcODE(args, x0, y0, ts, nt)
    plt.figure(figsize=(10, 10))
    plt.plot(sol[:, 1], sol[:, 0], 'r')
    
    args = (v + 0.1, w)
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(sol[:, 1], sol[:, 0], 'g')
    
    args = (v + 2, w)
    sol = calcODE(args, x0, y0, ts, nt)
    plt.plot(sol[:, 1], sol[:, 0], 'b')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()

    plt.figure(figsize=(10, 10))
    sol = calcODE(args, x0+100, y0+100, ts, nt)
    plt.plot(np.linspace(0, ts, nt), sol[:, 0], 'b')
    plt.plot(np.linspace(0, ts, nt), sol[:, 1], 'r')

    plt.xlabel('t')
    plt.ylabel('x, y')
    plt.grid()
    plt.show()
```
